```python
import os
import numpy as np
import SimpleITK as sitk
import pickle
import torch
from batchgenerators.utilities.file_and_folder_operations import join
from nnunetv2.inference.predict_from_raw_data import nnUNetPredictor
import nibabel as nib
import numpy as np
from scipy import ndimage
import pandas as pd
import os
import pickle

def process_nifti_file(nifti_file_path, binary_label_path, vmim, fpim):
    img = nib.load(nifti_file_path)
    data = img.get_fdata()
    data -= vmim
    data[data < 0] = 0
    
    label_data = fpim
    fpim[fpim<0.3] = 0
    # Count non-overlapping clusters
    overlap = (data.astype(int) & label_data.astype(int)).astype(bool)
    output_image_data = np.zeros_like(data)
   
    structure = np.ones((3, 3, 3), dtype=int)
    labeled_array, num_features = ndimage.label(data > 0, structure=structure)
         
    non_overlapping_mask = np.ones_like(data, dtype=bool)
    overlap_threshold = 50
    if num_features > 3:
        for cluster_id in range(1, num_features + 1):
            cluster_mask = labeled_array == cluster_id
            
            cluster_size = np.sum(cluster_mask)
            overlap_size = np.sum(cluster_mask & (label_data > 0))
            overlap_percentage = (overlap_size / cluster_size) * 100
            
            if overlap_percentage <= overlap_threshold:
                output_image_data[cluster_mask] = data[cluster_mask]
        
        new_img = nib.Nifti1Image(output_image_data, img.affine, img.header)
        nib.save(new_img, nifti_file_path)
    return output_image_data
    

def list_labels(directory):
    file_paths = []  # List to store the file paths
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            file_paths.append(file_path)
    return file_paths


def list_outputs(directory):
    file_paths = []  # List to store the file paths
    #     folds = ['fold_0', 'fold_1', 'fold_2', 'fold_3', 'fold_4', 'fold_5']
    entries = os.listdir(directory)
    folds = [entry for entry in entries if os.path.isdir(os.path.join(directory, entry))]
    for f in folds:
        subfolder = directory + f + '/validation'
        for root, dirs, files in os.walk(subfolder):
            for file in files:
                file_path = os.path.join(root, file)
                if file.endswith('.nii.gz') and not file.endswith('_softmax.nii.gz'):
                    file_paths.append(file_path)
    return file_paths

if __name__ == '__main__':
    vm = nib.load('Z:/Imaging/Multimodal/MRF/Peter/ventricular_mask.nii')
    vmim = vm.get_fdata()
    fp = nib.load('Z:/Imaging/Multimodal/MRF/Peter/nnUNet_CV/FP_prob.nii.gz')
    fpim = fp.get_fdata()    

    nifti_file_paths = list_outputs(
        'Z:/Imaging/Multimodal/MRF/Peter/nnUNet_CV/Dataset014_MRFzclin/nnUNetTrainer_250epochs__postprocessed2/')
    
    binary_label_paths = list_labels(
        'Z:/Imaging/Multimodal/MRF/Recon_MRF_3T/Patients/MRF_to_nnunet/nnUNet_raw/Dataset014_MRFzclin/labelsTr')
    img = nib.load('Z:/Imaging/Multimodal/MRF/Peter/header.nii')

    last_4 = {item[-16:] for item in nifti_file_paths}
    binary_label_paths = [item for item in binary_label_paths if item[-16:] in last_4]

    nifti_file_paths = sorted(nifti_file_paths, key=lambda x: x[-16:])
    binary_label_paths = sorted(binary_label_paths, key=lambda x: x[-16:])

    # print(nifti_file_paths)
    for nifti_file_path, binary_label_path in zip(nifti_file_paths, binary_label_paths):
        newimg = process_nifti_file(nifti_file_path, binary_label_path, vmim, fpim)
```

```python
import numpy as np
import nibabel as nib
import os
from matplotlib import pyplot as plt
import matplotlib
from sklearn.metrics import RocCurveDisplay, roc_curve, auc
import math

# This line of code is required for matplotlib to work properly.
matplotlib.use('QT5Agg')

# Constants.
base = 'Z:\\Imaging\\Multimodal\\MRF\\Peter\\nnUNet_CV\\Dataset016_MRFz\\'
tst_dir = base + 'nnUNetTrainer_250epochs__nnUNetPlans__3d_fullres_softmax\\'
softmax_dir = base + 'nnUNetTrainer_250epochs__nnUNetPlans__3d_fullres_softmax\\'
valnn_dir = base + 'valNN_probs\\'
labels_dir = 'Z:\\Imaging\\Multimodal\\MRF\\Recon_MRF_3T\\Patients\\MRF_to_nnunet\\nnUNet_raw\\Dataset016_MRFz\\labelsTr\\'
# brain_mask_dir = base + 'nnUNet_ROC\\'


# Initializations.
true_labels = []
pred_labels = []
valnn_labels = []

# Loop through patients in test set. Load brain mask (includes gray and white matter only), labels, and predicted labels.
file_list = os.listdir(tst_dir)
for file in file_list:
    if file.endswith('.nii.gz'):
        file_split_list = file.split('_')
        file_base = file_split_list[0] + '_' + file_split_list[1]
        label_img = nib.load(labels_dir + file_base + '.nii.gz')
        softmax_img = nib.load(softmax_dir + file_base + '_softmax.nii.gz')
        valnn_img = nib.load(valnn_dir + file_base + '_valnn.nii')
        # brain_mask_img = nib.load(brain_mask_dir + file_base + '_brain_mask.nii.gz')
        brain_mask_img = nib.load('Z:\\Imaging\Multimodal\\MRF\\Peter\\atlas_cortical.nii')
        label_img = label_img.get_fdata()
        softmax_img = softmax_img.get_fdata()
        valnn_img = valnn_img.get_fdata()
        valnn_img[valnn_img < 0] = 0
        valnn_img[valnn_img > 1] = 1
        brain_mask_img = brain_mask_img.get_fdata()
        # In order to make a proper ROC curve, extract voxels that are in the gray/white matter only for all
        # calculations going forward.
        label_img_masked = label_img[brain_mask_img == 1]
        softmax_img_masked = softmax_img[brain_mask_img == 1]
        valnn_img_masked = valnn_img[brain_mask_img == 1]
        true_labels.extend(label_img_masked.tolist())
        pred_labels.extend(softmax_img_masked.tolist())
        valnn_labels.extend(valnn_img_masked.tolist())

fpr, tpr, thresholds = roc_curve(true_labels, pred_labels)
fprv, tprv, thresholdsv = roc_curve(true_labels, valnn_labels)
roc_auc = auc(fpr, tpr)
roc_aucv = auc(fprv, tprv)
# Find optimal sensitivity and specificity.
dist_opt = 1
for idx in range(0, len(fpr)):
    dist = math.sqrt((1 - tpr[idx]) ** 2 + (0 - fpr[idx])**2)
    if dist < dist_opt:
        dist_opt = dist
        sens_opt = tpr[idx]
        spec_opt = 1 - fpr[idx]

print(sens_opt)
print(spec_opt)


# roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='MRF-DL model').plot()
plt.figure()
RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='MRF-DL model').plot()
RocCurveDisplay(fpr=fprv, tpr=tprv, roc_auc=roc_aucv, estimator_name='valNN by MAP18').plot()
plt.title('Voxel-level ROC')
plt.show()
# plt.savefig('Z:\\Imaging\\Multimodal\\MRF\\Peter\\nnUNet_CV\\Dataset016_MRFz\\ROC_Test_Set.png')
```
